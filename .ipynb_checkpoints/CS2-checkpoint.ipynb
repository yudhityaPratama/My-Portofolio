# Insight tentang Distribusi Kualitas Eksterior

Plot batang di atas menampilkan distribusi kualitas eksterior rumah dari dataset `house_categorical`. Berikut adalah beberapa insight berdasarkan plot:

1. Perbedaan Kualitas: Terlihat bahwa beberapa rumah memiliki kualitas eksterior yang sangat baik, sementara yang lain memiliki kualitas yang lebih rendah. Ini menunjukkan variasi yang signifikan dalam kualitas eksterior di antara rumah-rumah dalam dataset.
2. Tingkat Kualitas yang Paling Umum: Kualitas eksterior yang paling umum mungkin menunjukkan standar minimum atau preferensi umum dalam konstruksi atau renovasi rumah.
3. Potensi Korelasi: Kualitas eksterior rumah mungkin berkorelasi dengan faktor-faktor lain seperti harga jual, lokasi geografis, atau tahun pembangunan.
4. Perhatian pada Rencana: Melihat proporsi kategori kualitas eksterior tertentu dapat membantu perencanaan renovasi atau pembangunan baru dengan memahami preferensi pasar atau standar kualitas yang diharapkan.
5. Kesimpulan: Distribusi kualitas eksterior rumah ini memberikan wawasan tentang variasi dalam fitur fisik rumah yang dapat memengaruhi nilai dan daya tarik pasar.

Melalui pemahaman ini tentang kualitas eksterior, pemilik rumah atau pengembang dapat membuat keputusan yang lebih terinformasi tentang perawatan, renovasi, atau pengembangan properti mereka.


```python
x = house_numerical["SalePrice"]
sns.distplot(x, axlabel=x.name)
plt.title("Distribusi Harga Jual Rumah")
plt.xlabel('Harga Jual Rumah')
plt.ylabel('Frekuensi')
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

# Insight tentang Distribusi Harga Jual Rumah

Plot distribusi di atas menampilkan distribusi harga jual rumah dari data numerik dalam dataset. Berikut adalah beberapa insight berdasarkan plot:

1. Tren Sentral: Distribusi menunjukkan tren pusat yang kuat di sekitar nilai rata-rata dan median, yang mewakili titik-titik pusat sebaran harga jual rumah.
2. Skewness: Jika kurva distribusi miring ke kiri atau kanan, itu menunjukkan kecenderungan data menuju satu arah. Ini dapat memberikan wawasan tentang distribusi harga rumah yang mungkin tidak simetris.
3. Kesimpulan Statistik: Dengan nilai median, mean, dan standar deviasi yang tercetak di bawah plot, kita dapat mengevaluasi seberapa jauh data tersebar dari nilai rata-rata dan seberapa seragam distribusinya.
4. Potensi Outliers: Jika ada perbedaan yang signifikan antara nilai rata-rata dan median, atau jika standar deviasi besar, itu bisa menunjukkan adanya outlier (pencilan) dalam data harga jual rumah.

Melalui pemahaman ini tentang distribusi harga jual rumah, pemilik rumah atau pemangku kepentingan lainnya dapat mengambil keputusan yang lebih terinformasi tentang harga, strategi pemasaran, atau kebijakan harga. 


```python
x = house_numerical["GrLivArea"]
sns.distplot(x, axlabel=x.name)
plt.title("Distribusi Luas Hunian")
plt.ylabel('Frekuensi')
plt.xlabel('Luas Hunian (sqft)')
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

# Insight tentang Distribusi Luas Hunian

3. Rata-rata dan Median: Garis putus-putus menunjukkan rata-rata (hitam) dan median (merah), yang memberikan gambaran tentang pusat massa distribusi. Ini membantu untuk memahami apakah distribusi simetris atau tidak.
4. Statistik Deskriptif: Dalam statistik deskriptif di bawah grafik, kita melihat nilai median, rata-rata, dan standar deviasi. Ini memberikan pemahaman lebih lanjut tentang distribusi data.

Grafik ini dapat memberikan wawasan yang berguna bagi para pemangku kepentingan dalam memahami karakteristik luas hunian dalam dataset, yang dapat mempengaruhi penilaian atau keputusan terkait properti.


```python
x = house_numerical["PoolArea"]
sns.distplot(x, axlabel=x.name)
plt.title("Distribusi Luas Kolam Renang")
plt.xlabel('Luas Kolam Renang')
plt.ylabel('Frekuensi')
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

# Insight tentang Distribusi Luas Kolam Renang

Grafik distribusi di atas menampilkan distribusi data untuk luas kolam renang (`PoolArea`). Berikut adalah beberapa insight berdasarkan grafik dan statistik deskriptif:

1. Pusat Pemusatan: Terlihat bahwa nilai rata-rata (garis putus-putus hitam) dan median (garis putus-putus merah) memiliki perbedaan yang signifikan. Hal ini menunjukkan adanya potensi adanya pencilan (outlier) yang mempengaruhi rata-rata.
2. Distribusi Data: Distribusi data cenderung condong ke kanan (positif skewness), yang menunjukkan bahwa sebagian besar rumah mungkin memiliki kolam renang dengan luas yang relatif kecil.
3. Tren Pusat: Nilai median memberikan representasi yang lebih baik tentang "nilai tengah" data daripada rata-rata, terutama dalam kasus distribusi yang tidak simetris atau memiliki pencilan.
4. Penyebaran: Standar deviasi yang relatif tinggi menunjukkan variasi yang signifikan dalam luas kolam renang antar rumah.
5. Pertimbangan Renovasi: Informasi ini dapat berguna bagi pemilik rumah atau pengembang properti dalam merencanakan renovasi atau pembangunan yang melibatkan penambahan atau perbaikan kolam renang.

Melalui pemahaman ini tentang distribusi luas kolam renang, keputusan yang lebih baik dapat dibuat dalam perencanaan dan pengelolaan properti.


```python
x = np.log10(house_numerical["SalePrice"])
sns.distplot(x, axlabel=x.name, bins = 30)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.title("Distribution of Log10(SalePrice)")
plt.xlabel('Log10(SalePrice)')
plt.ylabel('Density')
plt.show()
```

# Insight tentang Distribusi Logaritma SalePrice

Plot histogram di atas menampilkan distribusi logaritma dari `SalePrice` dari dataset. Berikut adalah beberapa insight berdasarkan plot:

1. Tren Distribusi: Distribusi logaritma `SalePrice` membantu mengubah distribusi asimetris menjadi lebih simetris, membuatnya lebih sesuai untuk analisis statistik yang memerlukan asumsi normalitas.
2. Pusat Massa: Garis putus-putus yang menunjukkan rata-rata (hitam) dan median (merah) memberikan pandangan tentang di mana pusat massa data berada. Median lebih resisten terhadap pencilan dibandingkan rata-rata.
3. Kepadatan: Label sumbu-y yang dinamai "Density" menunjukkan bahwa sumbu y mewakili kepadatan distribusi, bukan frekuensi langsung. Ini memberikan informasi tentang seberapa sering nilai-nilai tertentu muncul dalam distribusi.
4. Skala: Label sumbu-x yang dinamai "Log10(SalePrice)" menunjukkan bahwa sumbu x mewakili nilai `SalePrice` yang telah diubah ke skala logaritma basis 10. Hal ini memungkinkan kita untuk memperhatikan perubahan dalam proporsi data pada skala yang lebih halus.
5. Kesimpulan: Distribusi logaritma memberikan wawasan tentang sebaran harga rumah dengan cara yang lebih terstruktur dan normal, yang memfasilitasi analisis statistik lebih lanjut.


```python
x = np.log10(house_numerical["GrLivArea"])
sns.distplot(x, axlabel=x.name, bins = 30)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.title("Distribution of Log10(GrLivArea)")
plt.xlabel("Log10(GrLivArea)")
plt.ylabel("Density")
plt.show()
```

### Insight

- Judul: Plot ini berjudul "Distribusi Log10(GrLivArea)", yang menunjukkan distribusi dari logaritma (basis 10) dari variabel "GrLivArea" (luas bangunan di atas tanah).
- Label Sumbu X: Sumbu x diberi label "Log10(GrLivArea)", yang merepresentasikan logaritma (basis 10) dari luas bangunan dalam kaki persegi.
- Label Sumbu Y: Sumbu y diberi label "Kepadatan", yang menunjukkan kepadatan kemunculan nilai-nilai dalam data.
- Garis Vertikal: Plot ini mencakup dua garis vertikal:
  - Garis putus-putus hitam mewakili rata-rata dari data yang telah ditransformasi dengan logaritma.
  - Garis putus-putus merah mewakili median dari data yang telah ditransformasi dengan logaritma.
- Statistik: 
  - Median dari data yang telah ditransformasi dengan logaritma adalah sekitar `x.median()`.
  - Rata-rata dari data yang telah ditransformasi dengan logaritma adalah sekitar `x.mean()`.
  - Standar deviasi dari data yang telah ditransformasi dengan logaritma adalah sekitar `x.std()`.
- Interpretasi: Transformasi logaritma membantu menormalkan distribusi "GrLivArea", mengurangi skewness dan membuatnya lebih mudah untuk dianalisis. Nilai rata-rata dan median yang dekat menunjukkan distribusi yang lebih simetris setelah transformasi.


```python
filtered_data = house_numerical[house_numerical["PoolArea"] > 0]
x = np.log10(filtered_data["PoolArea"])
sns.distplot(x, axlabel=x.name, bins = 30)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.title("Distribution of Log10(PoolArea)")
plt.xlabel("Log10(PoolArea)")
plt.ylabel("Frequency")
plt.show()
```
