---
### Masalah variable dengan NaN
* Untuk nilai NaN, ada beberapa _approach_ yang dapat dilakukan.
* Pertama kita dapat menghapus row yang memiliki nilai NaN.
    * Namun, hal ini tidak mungkin dilakukan, mengingat terdapat kolom yang 99% datanya adalah NaN.
    * Untuk kolom yang kurang dari 5% datanya adalah NaN, hal ini masih cukup layak untuk dilakukan.
* Kedua, kita dapat menghapus column yang memiliki nilai NaN.
    * Perlu diingat, bahwa menghapus kolom berarti menghapus fitur dari data.
    * Namun, untuk kolom seperti `PoolQC`, `MiscFeature`, `Alley`, dll. Hal ini dapat dilakukan.
    * Mengingat mayoritas dari variabel tersebut adalah NaN
* Ketiga, kita dapat melakukan _imputation_ atau mengisi nilai NaN dengan nilai seperti nilai _mean_.
    * Namun, untuk data yang mayoritas adalah NaN, hal ini tidak praktis.


---
* Untuk kolom yang mayoritas NaN, kita dapat membuang kolom tersebut.
* Untuk kolom yang memiliki sedikit nilai NaN, kita dapat melakukan _imputation_ atau membuang data-data tertentu.

## Numerical Data Visualization
* Untuk menganalisis distribusi persebaran data numerik dapat dilakukan dengan histogram.
* Sebagai contoh untuk analisis, akan digunakan variabel berikut:
    * SalePrice
    * LotArea
    * 2ndFlrSF

### SalePrice
* SalePrice merupakan variabel target yang ingin diprediksi.
* Secara fisik, ini berarti harga properti yang dijual.
* Karena secara fisik ini adalah harga jual berarti:
    * Nilai tidak bisa negatif.
    * Nilai pasti masuk akal: 
        * Rumah yang kualitasnya jelek, harganya juga akan murah
        * Akan ikut harga pasar di daerahnya. Misal bentuk dan ukuran yang sama, harga di Indonesia dan di US berbeda.

```python
x = house_numerical["SalePrice"]
sns.distplot(x, axlabel=x.name)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

**Analisis dan pertanyaan**
* Distribusi *right-skewed*.
* Kebanyakan data berada dalam range sekitar 100,000 sampai 300,000. Apakah ini wajar untuk harga rumah di US?
* Nilai selalu positif. Namun, kalau dimasukkan ke model ML hasil prediksi bisa jadi negatif kalau salah.
* Kita akan coba transformasi dengan log-scale.
    * Membuat distribusi terlihat seperti distribusi normal.
    * Memaksa nilai harga selalu positif, bahkan ketika dimasukkan kedalam model.

```python
x = np.log10(house_numerical["SalePrice"])
sns.distplot(x, axlabel=x.name)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

---
### 2ndFlrSF
* Beberapa variabel memiliki sparsity, contoh yang paling terlihat adalah `2ndFlrSF`.
* Variabel ini merepresentasikan luas area lantai 2 pada properti dalam satuan Square Foot.
    * Pada variabel ini terlihat bahwa banyak value yang berada di daerah nilai 0.
    * Lalu 0 sampai sekitar 200 tidak ada nilainya, dan lalu ada distribusi lagi yang muncul.
    * Sehingga akan sedikit sulit untuk dianalisis.

```python
x = house_numerical['2ndFlrSF']
sns.distplot(x, axlabel=x.name, bins = 20)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

* Kita dapat melihat bahwa cukup banyak nilai berada di angka dekat 0.
* Untuk ini akan kita cek apakah angka tersebut benar-benar 0 atau mendekati 0

```python
print('number of nonzero values:',np.sum(house_numerical['2ndFlrSF'] != 0))
print('number of zero values:',np.sum(house_numerical['2ndFlrSF'] == 0))
```

**Analisis**
* Ternyata data ini memiliki nilai 0 yang cukup banyak.
* Jika dilihat dari deskripsi, `2ndFlrSF: Second floor square feet`.
* Dari pernyataan tersebut kita dapat membuat suatu pernyataan.
* "Tidak semua rumah memiliki lantai kedua".
* Dari pernyataan tersebut kita dapat melakukan analisis lebih jauh lagi:
    * Untuk analisis lebih jauh, kita dapat membuat kluster dari data tersebut.
    * Yaitu kluster rumah dengan lantai kedua dan rumah tanpa lantai kedua

```python
x = house_numerical['2ndFlrSF'][house_numerical['2ndFlrSF'] != 0]
sns.distplot(x, axlabel=x.name, bins = 20)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

* Ketika kita mengambil kluster yang memiliki tingkat 2, maka distribusi variabel terlihat lebih baik.

### LotArea
* Beberapa variabel memiliki data anomaly, seperti `LotArea`.
* Variabel ini merepresentasikan luas tanah dari sebuah properti.
    * Banyak data terkumpul di daerah kiri distribusi.
    * Dan pada bagian kanan terdapat persebaran sedikit-sedkit.
    * Ini dapat menjadi dugaan anomali.

```python
x = house_numerical["LotArea"]
sns.distplot(x, axlabel=x.name, bins = 100)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

* Distribusi tersebut *right-skewed*
* Terdapat anomali dengan nilai yang sangat tinggi.
* Untuk analisis dapat dilakukan 2 pendekatan:
    * transformasi menjadi log-scale
    * drop outlier

**Transformasi log10**

```python
x = np.log10(house_numerical["LotArea"])
sns.distplot(x, axlabel=x.name, bins = 30)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

**Drop Outlier**

```python
x = house_numerical["LotArea"][house_numerical["LotArea"] < 20000]
sns.distplot(x, axlabel=x.name, bins = 30)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

* Dengan menggunakan transformasi skala maupun menghapus anomali distribusi menjadi lebih mendekati distribusi normal.
* Untuk analisis variabel-variabel lainnya, dapat dicoba sendiri.
* Namun perlu diingat, tujuan dari EDA adalah untuk eksplorasi.
* Sehingga output dari EDA adalah berupa pertanyaan dan hipotesis yang siap untuk diuji pada proses lebih lanjut.

## Pairplot
* Mengingat bahwa kita memiliki lebih dari 20 variabel numerik, tentu akan cukup tidak praktis untuk membuat plot satu per satu terhadap `SalePrice`.
* Untuk mempermudah pekerjaan tersebut, kita dapat menggunakan fitur `pairplot`

```python
subplots_col = 4
for i in range(0, len(house_numerical.columns), subplots_col):
    sns.pairplot(data=house_numerical,
                x_vars=house_numerical.columns[i:i+subplots_col],
                y_vars=['SalePrice'])
```
