#### T-test

```python
import scipy.stats as stats
```

A company wants to run a new campaign to increase the transactions’ amount of its product.
Before running the actual campaign to the whole customers, the company wants to know the campaign’s effectiveness in their 30 days transactions’ amount.
The average of transactions’ amount before the campaign is $100.
The company ran a testing on 20 customers and gathered their transactions’ amount in the next 30 days after the campaign. The data is stored in this table below.
Run a t-test analysis on the transactions’ amount average (with alpha = 5%). 


```python
#Import Data
data = np.array([100, 150, 50, 100, 130, 120, 100, 110, 75, 65, 150, 120, 50, 100, 100, 140, 90, 150, 50, 90])
data
```

```python
len(data)
```

```python
#Define H0 and H1
#H0: The average of the 30 days transactions amount = 100
#H1: The average of the 30 days transactions amount > 100
```

```python
#Define alpha = 0.05
```

```python
#Calculate t statistics and P-value
#perform one sample t-test

stats.ttest_1samp(a=data, popmean=100)
```

```python
#statistic=0.2731657711666385, pvalue=0.787672894169515
```

```python
#Conclude with alpha and p-value
# As the P-value is larger than alpha (5%), then we fail to the reject H0
# Conclusion: The campaign is not effective, the average of the 30 days transactions amount is still at 100
```

```python
#Conclude with t statistics and critical region
# Find the critical region
```

```python
#H1: larger (right side)
stats.t.ppf(1-0.05, 19)
```

```python
# Critical region: t > 1.73
# As the t-statistic is not in the critical region, then we fail to reject H0
# Conclusion:  The campaign is not effective, the average of the 30 days transactions amount is still at 100
```
