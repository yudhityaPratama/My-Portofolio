# Chapter 6

## 1. Show list of orders occurring from midnight before 12pm

```python
orders.head()
```

```python
query = """
select *
from orders
where strftime('%H', order_date) in ('00','01','02','03','04','05','06','07','08','09','10','11')
"""


sql_run(query)
```

## 2. Show a list of only the second last orders from each company

```python
query = """
select *, row_number() over(partition by company order by order_date desc) as rank
from orders
"""

sql_run(query)
```

```python
query = """
with tab as(
select *, row_number() over(partition by company order by order_date desc) as rank
from orders
)
select id,customer_id,order_id,order_date, company, status
from tab
where rank=2
"""

sql_run(query)
```

## 3. Count the orders from each company with the status REJECTED per day

```python
query = """
select company, strftime('%d', order_date) as tanggal, count(distinct order_id) as total_rejected
from orders
where status='REJECTED'
group by 1,2
"""

sql_run(query)
```

## 4. Group the orders by hour of order_date

```python
query = """
select strftime('%H', order_date) as hour_of_the_day, count(distinct order_id) as total_orders
from orders
group by 1
order by 1
"""

sql_run(query)
```

## 5. Group the orders by day and statuses as the example below :
### harus memuat kolom date, total_order_set, total_order_rejected

```python
query = """
select strftime('%d', order_date) as day_of_date, count(distinct case when status='SENT' then order_id end) as total_order_sent,
count(distinct case when status='REJECTED' then order_id end) as total_order_rejected
from orders
group by 1
"""

sql_run(query)
```

# Chapter 7

## 1. Show list of orders occurring in February 2020 with SENT status.

```python
query = """
select *
from orders
where order_date between '2020-02-01' and '2020-02-29' and status='SENT'
"""

sql_run(query)
```

## 2. Show a list of only the last orders from each company

```python
query = """
    select *, row_number() over(partition by company order by order_date desc) as rank
    from orders
"""

sql_run(query)
```

```python
query = """
with tab as (
    select *, row_number() over(partition by company order by order_date desc) as rank
    from orders
)

select id,customer_id,order_id,order_date, company, status
from tab
where rank = 1 -- last order
"""

sql_run(query)
```

## 3. Show a list of customers who made more than 1 SENT purchase

```python
query = """
select company, count(distinct case when status='SENT' then order_id end) as total_order_sent
from orders
group by 1
having total_order_sent>1
"""

sql_run(query)
```

## 4.Show the total orders and category of each company by following these criteria:
1. category harus : 
2. Good (total order status sent > 2 dan total rejected > 0), 
3. Superb (tidak ada yang di reject, dan total order status sent > 2), 
4. selain ini dia Normal

```python
query = """
select company,
count(distinct case when status='SENT' then order_id end) as total_order_sent,
count(distinct case when status='REJECTED' then order_id end) as total_order_rejected
from orders
group by 1
"""

sql_run(query)
```

```python
query = """
with cte as (
select company,
count(distinct case when status='SENT' then order_id end) as total_order_sent,
count(distinct case when status='REJECTED' then order_id end) as total_order_rejected
from orders
group by 1
)

select company,
case when total_order_sent>2 and total_order_rejected=0 then 'SUPERB'
      when total_order_sent>2 and total_order_rejected>0 then 'GOOD'
      else 'NORMAL'
end as category,
sum(total_order_sent + total_order_rejected) as total_orders
from cte 
group by 1,2
"""

sql_run(query)
```

## 1. Show the sum of the total value of the products shipped along with the Distributor Commissions (20% of the total product value if total quantity is 100 or less, 40% of the total product value if total quantity sold is more than 100)

```python
query = """
select Product_Name, quantity*price as Value,
round(case 
when quantity<=100 then (0.2*quantity*price) 
when quantity>100 then (0.4*quantity*price) end) as Distributor_Commission
from order_details
"""
sql_run(query)
```

## 2. Show total quantity of “Fanta(all variant)” shipped within February 2020

```python
query = """
select 'fanta all variant' as product, sum(quantity) as total_quantity
from order_details
where product_name like '%Fanta%'
"""
sql_run(query)
```

## 3. For each product, show the ID of the last order which contained that particular product

```python
query = """
select product_name, id, row_number() over(partition by product_name order by id desc) as rank
from order_details
"""
sql_run(query)
```

```python
query = """
select product_name,id from (
    select product_name, id, row_number() over(partition by product_name order by id desc) as rank
    from order_details
)
where rank=1 --last orders
order by id
"""
sql_run(query)
```

```python
query = """
with tab as(
select product_name, id, row_number() over(partition by product_name order by id desc) as rank
from order_details
)
select product_name,id from tab
where rank=1 --last orders
order by id
"""
sql_run(query)
```
